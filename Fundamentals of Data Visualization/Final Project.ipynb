Importning needed libraries and dataset.

```python
import pandas as pd
import altair as alt
import numpy as np
```

```python
data = pd.read_csv("World Happiness Report.csv")
```

```python
gdp = pd.read_csv("GDP per Year.csv")
```

```python
gdp.head()
```

```python
# Make a list of the wanted countries
countries = ['Argentina', 'Bolivia', 'Brazil', 'Chile', 'Colombia', 'Ecuador', 'Paraguay', 'Peru', 'Uruguay', 
             'Panama', 'Costa Rica', 'Nicaragua', 'Honduras', 'Guatemala', 'El Salvador']
```

```python
# Create a copy of the dataset with only the desired countries.
latam = data[data['Country Name'].isin(countries)].copy()
```

```python
gdp_latam = gdp[gdp['Country Name'].isin(countries)].copy()
```

```python
# Check that the countries wanted are actually there
print(latam['Country Name'].unique())
print(len(latam['Country Name'].unique()))
```

```python
print(gdp_latam['Country Name'].unique())
print(len(gdp_latam['Country Name'].unique()))
```

```python
gdp_latam.head()
```

```python
# Get the columns representing the years in the 'gdp_latam' DataFrame
year_columns = gdp_latam.columns[4:]  # Assuming the first two columns are 'Country Name' and 'Country Code'

# Filter the year columns to keep only those from 2005 to 2021
filtered_columns = [col for col in year_columns if int(col) >= 2005]

# Create a new dataset with the filtered columns
gdp_latam_filtered_dataset = gdp_latam[['Country Name', 'Country Code'] + filtered_columns].copy()

# Remove the last column from the 'gdp_latam_filtered_dataset' DataFrame
gdp_latam_filtered_dataset = gdp_latam_filtered_dataset.iloc[:, :-1]

# Display the updated dataset
gdp_latam_filtered_dataset

# Display the new dataset
gdp_latam_filtered_dataset
```

```python
# Pivot the 'gdp_latam_filtered_dataset' DataFrame
pivoted_dataset = gdp_latam_filtered_dataset.melt(id_vars=['Country Name', 'Country Code'],
                                                  var_name='Year',
                                                  value_name='Per Capita GDP')

# Display the pivoted dataset
pivoted_dataset
```

```python
# Define the dropdown selection
country_selection = alt.selection_single(
    name='Select',
    fields=['Country Name'],
    init={'Country Name': 'Argentina'},  # Set the initial selected country
    bind=alt.binding_select(options=list(pivoted_dataset['Country Name'].unique()))
)

# Create the Altair plot
plot = alt.Chart(pivoted_dataset).mark_line().encode(
    x='Year:N',
    y=alt.Y('mean(Per Capita GDP):Q', title='Average Per Capita GDP'),
    color=alt.condition(country_selection, 'Country Name:N', alt.value('lightgray')),
    opacity=alt.condition(country_selection, alt.value(1), alt.value(0.2)),
    tooltip=['Country Name', 'Year', 'Per Capita GDP']
).add_selection(
    country_selection
).properties(
    title='Average Per Capita GDP by Year per Country'
).configure_mark(
    size=2
).configure_view(
    width=600,  # Set the width of the plot
    height=400  # Set the height of the plot
)

# Display the plot
plot
```

```python
# Compute the average Per Capita GDP per country
avg_gdp = pivoted_dataset.groupby('Country Name')['Per Capita GDP'].mean().reset_index()

# Compute the average Perceptions Of Corruption per country
avg_corruption = latam.groupby('Country Name')['Perceptions Of Corruption'].mean().reset_index()

# Merge the average datasets
merged_data = pd.merge(avg_gdp, avg_corruption, on='Country Name')

# Create the Altair plot
plot = alt.Chart(merged_data).mark_circle().encode(
    y='Per Capita GDP:Q',
    x='Perceptions Of Corruption:Q',
    color='Country Name:N',
    size='Per Capita GDP:Q',
    tooltip=['Country Name', 'Per Capita GDP', 'Perceptions Of Corruption']
).properties(
    title='Average Per Capita GDP vs. Average Perceptions Of Corruption'
).interactive().configure_view(
    width=600,  # Set the width of the plot
    height=400  # Set the height of the plot
)

# Display the plot
plot
```

```python
# Compute the average Per Capita GDP per country
avg_gdp = pivoted_dataset.groupby('Country Name')['Per Capita GDP'].mean().reset_index()

# Compute the average Social Support per country
avg_social_support = latam.groupby('Country Name')['Social Support'].mean().reset_index()

# Merge the average datasets
merged_data = pd.merge(avg_gdp, avg_social_support, on='Country Name')

# Create the Altair plot
plot = alt.Chart(merged_data).mark_circle().encode(
    y='Per Capita GDP:Q',
    x='Social Support:Q',
    color='Country Name:N',
    size='Per Capita GDP:Q',
    tooltip=['Country Name', 'Per Capita GDP', 'Social Support']
).properties(
    title='Average Per Capita GDP vs. Average Social Support'
).interactive().configure_view(
    width=600,  # Set the width of the plot
    height=400  # Set the height of the plot
)

# Display the plot
plot
```

```python
# Group the 'latam' dataframe by 'Country Name' and calculate the average of 'Perceptions Of Corruption' and 'Log GDP Per Capita'
averages = latam.groupby('Country Name').agg({'Perceptions Of Corruption': 'mean', 'Log GDP Per Capita': 'mean'}).reset_index()

# Create the scatter plot with color encoding and tooltip
chart = alt.Chart(averages).mark_circle().encode(
    y=alt.Y('Log GDP Per Capita:Q', title='Average Log GDP Per Capita'),
    x=alt.X('Perceptions Of Corruption:Q', title='Average Perceptions of Corruption'),
    size=alt.Size('Log GDP Per Capita:Q', title='Average Log GDP Per Capita'),
    color=alt.Color('Perceptions Of Corruption:Q', title='Perceptions of Corruption',
                    scale=alt.Scale(scheme='blueorange')),  # Specify the color scale
    tooltip=['Country Name']  # Add tooltip for country name
).properties(
    title='Average Perceptions of Corruption vs Average Log GDP Per Capita per Country'
).interactive()  # Enable zooming in and out

# Display the chart
chart
```

```python

# Pivot the 'latam' dataframe to have years as columns and 'Social Support' as values
pivoted_data = latam.pivot(index='Country Name', columns='Year', values='Social Support').reset_index()

# Reshape the pivoted data into a tidy format
tidy_data = pivoted_data.melt(id_vars='Country Name', var_name='Year', value_name='Social Support')

# Create the line chart with dots and tooltips
chart = alt.Chart(tidy_data).mark_line().encode(
    x='Year:O',
    y='Social Support:Q',
    color='Country Name:N'
).properties(
    title='Change in Social Support Year by Year',
    width=600,
    height=400
).interactive()

# Add medium-sized dots for each data point
dots = alt.Chart(tidy_data).mark_circle(size=50).encode(
    x='Year:O',
    y='Social Support:Q',
    color='Country Name:N',
    tooltip=['Country Name', 'Year', 'Social Support']
)

# Combine the line chart and dots
chart = chart + dots

# Display the chart
chart
```

```python
# Group the 'latam' dataframe by 'Country Name' and calculate the average of 'Social Support' and 'Freedom To Make Life Choices'
averages = latam.groupby('Country Name').agg({'Social Support': 'mean', 'Freedom To Make Life Choices': 'mean'}).reset_index()

# Create the scatter plot with color encoding
chart = alt.Chart(averages).mark_circle().encode(
    y=alt.Y('Freedom To Make Life Choices:Q', title='Average Freedom To Make Life Choices'),
    x=alt.X('Social Support:Q', title='Social Support'),
    size=alt.Size('Freedom To Make Life Choices:Q', title='Average Freedom To Make Life Choices'),
    color=alt.Color('Social Support:Q', title='Social Support',
                   scale=alt.Scale(scheme='blueorange')),
    tooltip=['Country Name']  # Add tooltip for country name
).properties(
    title='Average Social Support vs Average Freedom To Make Life Choices per Country'
).interactive()

# Display the chart
chart
```

```python
# Pivot the 'latam' dataframe to have years as columns and 'Freedom To Make Life Choices' as values
pivoted_data = latam.pivot(index='Country Name', columns='Year', values='Freedom To Make Life Choices').reset_index()

# Reshape the pivoted data into a tidy format
tidy_data = pivoted_data.melt(id_vars='Country Name', var_name='Year', value_name='Freedom')

# Create a selection dropdown for country
country_dropdown = alt.binding_select(options=list(tidy_data['Country Name'].unique()))
country_selection = alt.selection_single(fields=['Country Name'], bind=country_dropdown, name='Country')

# Create the line chart
line_chart = alt.Chart(tidy_data).mark_line().encode(
    x='Year:O',
    y='Freedom:Q',
    color=alt.condition(country_selection, 'Country Name:N', alt.value('lightgray'), legend=None),  # Add legend=None
    opacity=alt.condition(country_selection, alt.value(1), alt.value(0.2))  # Fade out non-selected lines
).properties(
    title='Change in Freedom to Make Life Choices per Year',
    width=600,
    height=400
).add_selection(country_selection)

# Display the chart
line_chart
```
