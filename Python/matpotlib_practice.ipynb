```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
x = np.arange(10)
y = np.arange(0, 1000, 100)
```

```python
plt.plot(x,y)
```

```python
x = np.random.normal(0.5, 0.1, (100))
x.sort()
y = x**2
print(x)
print(y)
plt.plot(x, y)
```

```python
plt.plot(x, y, 'o') # instead of plotting a line, you can plot the specific points.
```

```python
x = np.arange(100)
y = x**3
plt.plot(y, '1') #This changes the representations in the plot.
```

```python
#Different types of markers in Matplotloib
markers = ['o', '*', '.', ',', 'x', 'X', '+', 'P', 's', 'D', 'd', 'p', 'H', 'h', 'v', '^', '<', '>', '1', '2', '3', '4', '|', '_']

for m in markers:
    print(m)
    plt.plot(y, m)
    plt.show()
```

```python
y = np.array([1, 3, 13]) 
line = ['-',':','--','-.'] # Different types of lines in Mtplotlib
for l in line:
  print(l)
  l = 'o' + l + 'r'
  plt.plot(y, l)
  plt.show()
```

```python
color = ['r','g','b','c','m','y','k','w'] # Colors in matplotlib
for c in color:
    print(c)
    c = 'o:' + c
    plt.plot(y, c)
    plt.show()
```

```python
plt.plot(y, marker = '^', ms = 30) # Changing the size of the markers
```

```python
# Plotting multiple lines in the same plot
x = np.arange(10)
y1 = x*2
y2 = x**2
plt.plot(y1)
plt.plot(y2)
```

```python
plt.plot(x, y1)
plt.plot(x, y2)
```

```python
# Different presentations, labels and grid
plt.plot(x, y1, label = 'double')
plt.plot(x, y2, label = 'square')
plt.xlabel("Xs")
plt.ylabel("Ys")
plt.grid()
plt.legend(['1st', '2nd'])
```

```python
# Sub Plots one on top of the other
plt.subplot(2, 1, 1)
plt.plot(x, y1, label = 'double')
plt.legend()
plt.xlabel("Xs")
plt.ylabel("Ys")
plt.subplot(2, 1, 2)
plt.plot(x, y2, label = 'square')
plt.legend()
plt.xlabel("Xs")
plt.ylabel("Ys")
plt.grid()
```

```python
# Sub plots side by side

plt.subplot(1, 2, 1)
plt.plot(x, y1, label = 'double')
plt.legend()
plt.xlabel("Xs")
plt.ylabel("Ys")
plt.subplot(1, 2, 2)
plt.plot(x, y2, label = 'square')
plt.legend()
plt.xlabel("Xs")
plt.ylabel("Ys")
plt.grid()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100]) # Color Bar

plt.scatter(x, y, c=colors, cmap='hot') # Cmap for color scheme

plt.colorbar()
```

```python
# Plots with size relations

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes)
```

```python
# Plots with size relation and color

plt.scatter(x, y, c = colors, cmap = 'hot', alpha = 0.5, s = sizes)
plt.colorbar()
```

```python
# Bar plots
plt.bar(x, y)
```

```python
# Histograms
plt.hist(x)
```

```python
# Pir charts
x = [0.66, 0.25, 0.09]
plt.pie(x, labels = ['2/3', '1/5', 'others'])
plt.legend()
```

```python
# More interesting pie charts
x = [0.66, 0.25, 0.09]
plt.pie(x, labels = ['2/3', '1/5', 'others'], explode = [0.0, 0.2, 0])
plt.legend()
```

```python
# More than two lines in the same plot

x = np.arange(10)
alpha = (x+1)**2/2
delta = (x+2)**2
omicron = (x+3)**2 * 3

plt.plot(x, alpha, label = "Alpha")
plt.plot(x, delta, label = "Delta")
plt.plot(x, omicron, label = "Omicron")
plt.legend()
```
